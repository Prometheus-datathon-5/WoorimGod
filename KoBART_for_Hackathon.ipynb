```python

model = BartForConditionalGeneration.from_pretrained('./kobart_summary')

model.eval()
```

```python
tokenizer = PreTrainedTokenizerFast.from_pretrained('digit82/kobart-summarization')
```

```python
cd /content/drive/MyDrive/prometheus/kobart
```

```python
test_dataset = pd.read_csv("test.csv", sep=",")
test_dataset.drop('id', axis=1, inplace=True)
print(test_dataset)
```

```python
print(test_dataset.info())
```

```python
from konlpy.tag import Okt


okt = Okt()
preds = []

for i in range(len(test_dataset)):
  raw_input_ids = tokenizer.encode(test_dataset.iloc[i][0], max_length=510, truncation=True)
  input_ids = [tokenizer.bos_token_id] + raw_input_ids + [tokenizer.eos_token_id]
  summary_ids = model.generate(torch.tensor([input_ids]),  num_beams=4,  max_length=512,  eos_token_id=1)
  summary_text = tokenizer.decode(summary_ids.squeeze().tolist(), skip_special_tokens=True)
  preds.append(summary_text)


preds = [okt.morphs(pred) for pred in preds]
```

```python
new_preds = pd.Series(preds)
new_preds = pd.DataFrame(new_preds)
new_preds.reset_index(inplace=True)

new_preds.rename(columns={'index':'id', 0:'pred'}, inplace=True)
```
